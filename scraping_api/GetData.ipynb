```python
response_json = response.json()
print(len(response_json['data']))
```

```python
def prettyPrint(obj):
    print(json.dumps(obj, indent=True))

prettyPrint(response_json)
```

```python
for attribs, values in response_json.items():
    print(attribs)
```

```python
response_json['links']
```

```python
responses = []

while 'next' in response_json['links']:
    responses.append(response_json)
    url = response_json['links']['next']
    response = requests.get(url, headers=headers, params={"page[size]": "100"})
    response_json = response.json()
responses.append(response_json)
```

```python
import pandas as pd

responses_df = pd.DataFrame(responses)
```

```python
responses_df.head()
```

```python
responses_df.to_csv('./responses.csv', index=False)
```

```python
reviews = []
for response_list in responses_df['data']:
    reviews.extend(response_list)

print(len(reviews))
```
